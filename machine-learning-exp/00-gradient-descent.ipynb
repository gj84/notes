```python
import matplotlib.pyplot as plt

def ploter(dataset, linear=None):
    xs = list(map(lambda x: x[0], dataset))
    ys = list(map(lambda y: y[1], dataset))

    plt.plot(xs, ys, 'go')
    if linear:
        lx = [x[0] for x in linear]
        ly = [y[1] for y in linear]
        plt.plot(lx, ly, 'c-')
    plt.ylabel('Gradient Descent')
    plt.show()
```

```python
dataset = [(1,1),
           (2,1),
           (5,1),
           (6,3),
           (8,2),
           (9,5),
           (10,10),
           (13,4),
           (15,10),
           (17,11),
           (21,12),
           (22,17),
           (23,18),
           (26,20),
           (28,28),
           (30,25),
           (31,26),
           (33,27),
           (34,32),
           (35,28),
           (38,32),
           (40,34)]
```

```python
with open('data', 'w', encoding="ascii") as f:
    for d in dataset:
        dato = "%d  %d\n" % (d[0], d[1])
        print(dato)
        f.write(dato)
```

```python
class Hipotesis: #simple linear function
    def __init__(self, theta0, theta1): #linear
        self.theta0 = theta0
        self.theta1 = theta1
        
    def __str__(self):
        return ("y: %f + %f x" % (self.theta0, self.theta1))

    def get_values(self):
        return (self.theta0, self.theta1)

    def new_values(self, **kwargs):
        if "theta0" in kwargs:
            self.theta0 = kwargs["theta0"]
        
        if "theta1" in kwargs:
            self.theta1 = kwargs["theta1"]

    def evaluate(self, x):
        return x * self.theta1 + self.theta0
    
    def eval_list(self, lis):
        return [(x, self.evaluate(x)) for x in lis]
```

```python
class Gradient:
    def __init__(self, dataset, hipotesis, learn, precision):
        self.dataset = dataset
        self.m = len(dataset)
        self.h = hipotesis
        self.precision = precision
        self.learn = learn
        self.lastth1 = hipotesis.theta1 - (precision * 10)
        self.costs = []
        
    def cost_function(self):
        su = 0
        for x, y in self.dataset:
            su += (self.h.evaluate(x) - y)**2
        self.costs.append(su / float(2.0 * len(dataset)))

    def convergence(self):
        if (self.h.theta1 - self.lastth1) <= self.precision: #mean the slope's change
            print(self.h)
            return True
        return False
    
    def apply(self):
            
        error_sum0 = sum([(self.h.evaluate(el[0]) - el[1]) for el in self.dataset])
        theta0 = self.h.theta0 - (self.learn * error_sum0  / self.m) 
        
        error_sum1 = sum([(self.h.evaluate(el[0]) - el[1]) * el[0] for el in self.dataset])
        theta1 = self.h.theta1 - (self.learn * error_sum1 / self.m)
        
        self.lastth1 = self.h.theta1 
        
        self.h.new_values(theta0=theta0, theta1=theta1)
        self.cost_function()
```

```python
hipotesis = Hipotesis(0, 0) #initial
gradient = Gradient(dataset, hipotesis, .0008, precision=.00006)

k = 0
while not gradient.convergence():
    linear = gradient.h.eval_list(range(41))
    ploter(dataset, linear)
    gradient.apply()

```
